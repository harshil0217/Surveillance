```python
def drop_num(x):
    entry = x
    if (type(entry) != float):
        entry = entry.split(" | ")
        if (len(entry) > 1):
            entry = entry[1]
    return entry
```

```python
def remove_num(data, col_names):
    for col_name in col_names:
        col = data[col_name]
        n_col = col.apply(lambda x: drop_num(x))
        data[col_name] = n_col
    return data
    
```

```python
def make_year_country_table(data, year_col, country_col):
    yc = pd.DataFrame(columns = [year_col, country_col])
    years = np.unique(data[year_col])
    countries = np.unique(data[country_col])
    index = 0
    for year in years:
        for country in countries:
            yc.loc[index] = [year, country]
            index += 1
    return yc
            
```

```python
def get_counts(x, count_data, infra_col, year_col, country_col):
    country = x[country_col]
    year = x[year_col]
    filtered = count_data.loc[count_data[year_col] == year]
    filtered = filtered.loc[filtered[country_col] == country]
    values = filtered[infra_col].dropna()
    values = values.value_counts()
    for index in values.index:
        x[index] = x[index] + values[index]
    return x
    
    
```

```python
def counts(instance_data, count_data, infra_col, year_col, country_col):
    instance_data = instance_data.apply(lambda x: get_counts(x, count_data, infra_col, year_col, country_col), axis = 1)
    return instance_data
```

```python
def Between(x, year_col, begin, end):
    if int(x[year_col]) >= begin and int(x[year_col]) <= end:
        return True
    else:
        return False
    
def YearCutoff(data, year_col, begin, end):
    locs = data.apply(lambda x: Between(x, year_col, begin, end), axis = 1)
    data = data[locs]
    data = data.reset_index(drop = True)
    return data
```

```python
def DropCountriesRow(x, countries, country_col):
    if (x[country_col] in countries):
        return False
    else:
        return True
    
```

```python
def DropCountries(data, countries, country_col):
    locs = data.apply(lambda x: DropCountriesRow(x, countries, country_col), axis = 1)
    data = data[locs]
    data = data.reset_index(drop = True)
    return data
```

```python
def ExcludeStates(x, country_col, states, keep):
    if (keep == True):
        if (x[country_col] in states):
            return True
        return False
    else:
        if (x[country_col] in states):
            return False
        return True

def RemoveStates(data, country_col, states, keep = False):
    keep = data.apply(lambda x: ExcludeStates(x, country_col, states, keep), axis = 1)
    data = data[keep]
    data = data.reset_index(drop = True)
    return data
```

```python
def ApplyIndex(x, index_data, col_range, year_country_col):
    year_country = x[year_country_col]
    filtered = index_data.loc[index_data[year_country_col] == year_country, 
                              index_data.columns[col_range[0]:col_range[1]]]
    filtered = pd.Series(filtered.iloc[0])
    return filtered
```

```python
def AddIndexCols(data, index_data, col_range, year_country_col):
    newcols = data.apply(lambda x: ApplyIndex(x, index_data, col_range, year_country_col), axis = 1, 
                         result_type = "expand")
    data = pd.concat([data, newcols], axis = 1)
    return data
    
    
    
    
```
